```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
csv_raw = pd.read_csv("../data/raw/comments_train.csv")
csv_clean = pd.read_csv("../data/processed/comments_clean.csv")
```

```python
csv_raw
```

```python
idx_comm = csv_raw.comment.apply(lambda x: len(x)).sort_values(ascending=False).index
print(csv_raw.iloc[idx_comm].comment.values[0])
print("CLEAN")
print(csv_clean.iloc[idx_comm].x.values[0])
```

```python
wc = WordCloud().generate(" ".join(csv_raw.comment.values))
wc_clean = WordCloud().generate(" ".join(csv_clean.x.values))

fgi, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
ax1.imshow(wc, interpolation="bilinear")
ax1.axis("off")

ax2.imshow(wc_clean, interpolation="bilinear")
ax2.axis("off")
plt.show()
```
